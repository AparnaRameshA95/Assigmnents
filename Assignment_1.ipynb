```python
iris_data=iris_data.rename(columns={'SL':'Sepal_length','SW':'Sepal_width',
                                    'PL':'Petal_length','PW':'Petal_width'})
iris_data
```

```python
import matplotlib.pyplot as plt
```

```python
plt.xlabel('Petal_length')
plt.ylabel('Petal_width')
plt.title('Bar Graph of Petal Length vs Petal Width')
plt.bar(iris_data['Petal_length'],iris_data['Petal_width'])
```
